```python
#recommender system is a piece of code which is intelligent enough to understand user preferences and suggest new contents accordingly.
#Types of recommender system:

# Content based
# Collaborative filtering
# Hybrid


# Content-based filtering recommends depending on the similarities of the contents.
# Suppose you watch too much anime contents, then you will be recommended more anime content,
#fewer songs, or other contents. Here tags of things are created and based on the similarities of those tags recommendation is being done.
#Startups at the beginning use this filtering.

# Collaborative filtering works on users’ interests. 
# Let's say user A and user B, whose behavior is very similar(similarity score 0.9).
# After watching a movie, they review or rate the movie in the same way.
# So if I show a movie to A and if A likes the movie, then I can recommend the movie to B too.
# Facebook uses this type of filtering.

# A hybrid is a combination of both above. YouTube was using content-based filtering first, but now it is thought that it is using hybrid filtering.
#The types of videos I regularly watch will be shown to me, also content that is viewed by people like me will also be recommended to me.

# So we are thinking of using a content-based recommender system in this project.

# Project Flow:
# We will start with data.
# We will perform preprocessing.
# Build a model.
# Use it in a website.
# Deploy the model.
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3: #it will loop for 3 times and 3 names will be fetched
            L.append(i['name'])
        counter+=1
    return L 
```

```python
#as we can see it is also a list of dictionaries
#we need those dictionarires where the job is director
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
#So...Now we want to remove spaces(blank spaces with no spaces) from the words else it could make different senses.
#movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
#we could do this like this or we can use a helper function
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#So we have to calculate similarities between these tags
#all of it is textual data
#So we have to use text vectorization
#turn these tags to vector
#vectors has co ordinates....X and Y
#So if user puts a movie name then we will first take that movies vector
#And recommend the closest vectors related movie names
#we will use bag of words[there are other techiques like tfidf word2vec]
#first concatinate all the tags
#then extract top 5000 words whose frequency is high
#then we will pick one movie's tag and see if any of the top 5000 words occured here
#if it occured then it occured for how many times

#         w1| w2| w3....| w5000 //w1=word 1.....w2=word 2
#muvi1    5    3   1       0    //number of occurance
#muvi2    1    3   0       1
#muvi3    0    1   7       4
#muvi5000 4    2   8       1

#shape(5000,5000)
#now each movie is a vector
#during this vectorization proecess we will not consider stopwords(in or to is)
#count vectorizer class will be used 
#then we will use cosine similarities .... 
#Vectors which have smaller angle difference between them they are more similar
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
 import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('danced')
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
#similarities will be calculated for all to all ... that means total 4806x4806
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Aliens')
```
